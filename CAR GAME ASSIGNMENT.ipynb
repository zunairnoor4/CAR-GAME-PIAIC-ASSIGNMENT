```python
# CAR GAME

# PERSONAL INFORMATION

        '''Name: ZUNAIR NOOR
        
        Roll Number: AIC026723
        
        Teacher Name: Noman Islam
        
        Institute Name:Saylani Office Head Campus
        
        Timing:Sunday (9AM to 12PM)'''
    


''' MUST GO THROUGH THE CODE IT IS ABSLUTELY AMAZING ALSO MUST OPERATE IT IT IS A MUCH FUN'''

import IPython.display
print('                                *******************')
print('                                  A.B.C Car game')
print('                                *******************')
print('')
print('                            Welcome to the A.B.C Car game')
print('')
print('If you want some help, simply type "help"')

operation = ""
started= False                     # defining boolean to not start or stop the car twice
gear = 0

while True:
    operation = input(">")
    if operation.lower()== "help":
        IPython.display.clear_output()
        print('start (to start the car)')
        print('gear up (to increase the gear)')
        print('gear down (to decrease the gear)')
        print('stop (to stop the car)')
        print('quit (to exit the game)')
        print('')
        
    elif operation.lower() == 'start':
        if started:              # READ AS "if started is true then"
            IPython.display.clear_output()
            print('Hey, the Car is already started')
            print('')
        else:
            IPython.display.clear_output()
            print('lets go the car is started...')
            started = True
            print('')
       
    
    elif operation.lower()=="stop":
        if gear == 0:            
            if not started:                      # READ AS " if started is not true then."
                IPython.display.clear_output()
                print('Hey, the Car is already stopped')
                print('')
            else:
                IPython.display.clear_output()
                print('the car is stopped!!!')
                print('')
                started = False
        else:
            IPython.display.clear_output()
            print(f'your car is currently at {gear}, first neutral the car and then try to stop it... ')
            print('')
            
            
    elif operation.lower() == 'gear up':
        if started:
            if gear ==0:
                gear += 1
                IPython.display.clear_output()
                print(f'your gear is {gear}')
                print('the car is about to move...')
                print('')
            elif gear >= 6:
                IPython.display.clear_output()
                print(f'your gear is already {gear}')
                print("maximum six gears are there in car...")
                print('')
            else:
                gear += 1
                IPython.display.clear_output()
                print(f'your gear is {gear}')
                print("gear is up, moving more faster now...")
                print('')
        else:
            IPython.display.clear_output()
            print('Hey! first start the car then gear up...')
            print('')
            
        
    elif operation.lower() == 'gear down':
        if started:
            if gear ==0:
                IPython.display.clear_output()
                print(f'your gear is already {gear}')
                print('the car is at neutral ...')
                print('')
            elif gear == 1:
                gear -= 1
                IPython.display.clear_output()
                print(f'your gear is {gear}')
                print("the car is neutral now...")
                print('')
            else:
                gear -= 1
                IPython.display.clear_output()
                print(f'your gear is {gear}')
                print("gear is down, moving more slower now...")
                print('')
        else:
            IPython.display.clear_output()
            print('Hey! first start the car then gear down...')   
            print('')    
    
    
    elif operation.lower()=="quit":
        IPython.display.clear_output()
        print('bye...')
        break
    
    else:
        IPython.display.clear_output()
        print('sorry i donot understand you')
```
